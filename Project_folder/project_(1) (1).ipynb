### Outlier Replacement

```python
df.median()
```

```python
q1 = df.BMI.quantile(0.25) #Q1
q3 = df.BMI.quantile(0.75) #Q3
```

```python
print(q1)
print(q3)
```

```python
IQR=q3-q1
IQR
```

```python
upper_limit = q3+1.5*IQR
upper_limit
```

```python
lower_limit = q1-1.5*IQR
lower_limit
```

```python
df['BMI'] = np.where(df['BMI']>upper_limit,27.0,df['BMI'])
df['BMI'] = np.where(df['BMI']<lower_limit,27.0,df['BMI'])
```

```python
sns.boxplot(df.BMI)
```

```python
q1 = df.GenHlth.quantile(0.25) #Q1
q3 = df.GenHlth.quantile(0.75) #Q3
```

```python
print(q1)
print(q3)
```

```python
IQR=q3-q1
IQR
```

```python
upper_limit = q3+1.5*IQR
upper_limit
```

```python
lower_limit = q1-1.5*IQR
lower_limit
```

```python
df['GenHlth'] = np.where(df['GenHlth']>upper_limit,2.0,df['GenHlth'])
# df['GenHlth'] = np.where(df['GenHlth']<lower_limit,2.0,df['GenHlth'])
```

###x and y Split


```python
x=df.drop(columns=['Diabetes_binary'],axis=1)
x.head()
```

```python
y=df['Diabetes_binary']
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scale =MinMaxScaler()
```

```python
x_scaled=pd.DataFrame(scale.fit_transform(x),columns=x.columns)
x_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.3,random_state=0)
```

###**Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model1 = DecisionTreeClassifier()
```

```python
model1.fit(x_train,y_train)
```

```python
y_predict = model1.predict(x_test)
```

```python
y_predict
```

```python
y_predict_train = model1.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print('Testing Accuracy = ', accuracy_score(y_test,y_predict))
print('Training Accuracy = ', accuracy_score(y_train,y_predict_train))
```

```python
pd.crosstab(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```
